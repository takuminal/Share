```python
import customtkinter
from modules import function as fc
from rdkit import Chem
import pubchempy as pcp
import csv
import pandas as pd



FONT_TYPE = "meiryo"

class App(customtkinter.CTk):

    def __init__(self):
        super().__init__()

        # メンバー変数の設定
        self.fonts = (FONT_TYPE, 15)
        # フォームサイズ設定
        self.geometry("700x400")
        self.title("Basic GUI")

        # フォームのセットアップをする
        self.setup_form()
    
    def setup_form(self):
        # CustomTkinter のフォームデザイン設定
        customtkinter.set_appearance_mode("dark")  # Modes: system (default), light, dark
        customtkinter.set_default_color_theme("blue")  # Themes: blue (default), dark-blue, green

        #入力
        self.textbox1 = customtkinter.CTkEntry(
            master=self, 
            placeholder_text="Input Chemical Name", 
            width=200, 
            font=self.fonts)
        self.textbox1.grid(row=0, column=0, padx=10, pady=20)
        
        #実行ボタン
        self.button1 = customtkinter.CTkButton(
            master=self, text="Run", 
            command=self.button_function, 
            font=self.fonts)
        self.button1.grid(row=0, column=1, padx=10, pady=20)

        #実行ボタン
        self.output = customtkinter.CTkLabel(
            master=self, text="Run", 
            font=self.fonts)
        self.output.grid(row=1, column=1, padx=10, pady=20)
    

    def button_function(self):
        # テキストボックスに入力されたテキストを表示する
        
        input_val = self.textbox1.get()
        results = pcp.get_compounds(input_val, 'name')
        compound = results[0]
        smiles = compound.canonical_smiles
        fg = fc.count_functional_groups(smiles)
        self.output.update()
        


if __name__ == "__main__":
    # アプリケーション実行
    app = App()
    app.mainloop()
```
