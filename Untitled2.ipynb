```python
import tkinter as tk
from tkinter import filedialog

def select_csv_file():
    root = tk.Tk()
    root.withdraw()  # メインウィンドウを表示しないようにする

    file_path = filedialog.askopenfilename(
        title="Select CSV File",
        filetypes=[("CSV files", "*.csv"), ("All files", "*.*")]
    )

    if file_path:
        return file_path
        # ここで選択されたファイルパスを使って何かを行う

```

```python

if __name__ == "__main__":
    select_csv_file()
```

```python
file_path = select_csv_file()
```

```python
import pandas as pd 
```

```python
df = pd.read_csv(file_path,encoding = "shift-jis")
# 行を逆にする
df_2 = df[::-1]



#日付をdatetime型に変える関数
def time_edit(text):
    from datetime import datetime
    # カンマで文字列を分割し、リストの要素として取得
    text2 = text.split('/')
    text3 = text2[2].split(" ")
    text4 = text3[1].split(":")
    
    t_info = text2[:-1]+text3[:-1]+text4
    yr = int(t_info[0])
    mnt = int(t_info[1])
    day = int(t_info[2])
    hr = int(t_info[3])
    min = int(t_info[4])
    return datetime(yr,mnt,day,hr,min)

def benefit_edit(text):
    if text !="-":
        out = float(text)
    else:
        out = 0
    return out

def sumup(ls):
    out =[]
    mass= 0
    
    for elem in ls:
        mass =  mass + elem
        out.append(mass)
    return(out)
        


df['注文日時'] = df['注文日時'].apply(time_edit)
df['約定日時'] = df['約定日時'].apply(time_edit)
df["取引損益"] = df["取引損益"].apply(benefit_edit)
```
