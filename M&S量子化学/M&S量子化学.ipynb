### 黒体輻射
  温度により物体が光を発する現象であり、温度帯により波長が異なる。この波長温度依存性の関係は、古典物理では説明できなかった。古典物理の下で導かれる黒体輻射は、レイリー―ジーンズによる式(1)で表現された。
$$
d\rho(\nu, T) = \rho(T)d\nu = \frac{{8 \pi k_BT}}{{c^3}} \cdot \nu^2 d\nu$$

$\rho$ : エネルギー密度
$T$　: 温度, 
$\nu$　: 振動数, 
$k_B$　: ボルツマン定数, 




```python
def Rayleigh_Jeans(nu,T):
    import math

    d_nu = 1*10**14
    
    pi = math.pi
    kB = 1.38*10**(-23)
    c = 2.998*10**9
    
    return (8*pi*kB*T)*(nu**2)*d_nu/(c**3)
```

```python
import numpy as np
nu_ls = np.linspace(0,20e14,100)
T_ls =[2000, 3000, 5000,20000]


import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))

for T in T_ls:
    rho_ls = [Rayleigh_Jeans(nu,T) for nu in nu_ls]
    plt.plot(nu_ls,rho_ls,label= f"{T}K")
    
plt.legend()
```

```python
def Plank(nu,T):
    import math

    d_nu = 0.01*10e14
    
    pi = math.pi
    kB = 1.38*10e-23
    c = 2.998*10e9
    h = 6.626*10e-34
    E = math.exp((h*nu)/(kB*T))
    
    return (8*pi*h)*(nu**3)*d_nu/((c**3)*E-1)
```

```python
import numpy as np
nu_ls = np.linspace(0,20e14,100)
T_ls =[2000, 3000, 5000,10000]


import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))

for T in T_ls:
    rho_ls = [Plank(nu,T) for nu in nu_ls]
    plt.plot(nu_ls,rho_ls,label= f"{T}K")
    
plt.legend()
```
